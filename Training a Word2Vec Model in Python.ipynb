```python
## loading and processing the data 
import nltk
from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec
import string
from numpy import triu



# Download necessary data
nltk.download('punkt')

# Sample training corpus
corpus = [
    "Machine learning is fascinating.",
    "Deep learning is a subset of machine learning.",
    "Word embeddings capture semantic meaning.",
    "Natural Language Processing is a field of AI.",
    "Transformers have revolutionized NLP."
]

# Tokenize sentences
tokenized_corpus = [word_tokenize(sentence.lower()) for sentence in corpus]

# Remove punctuation
tokenized_corpus = [[word for word in sentence if word not in string.punctuation] for sentence in tokenized_corpus]
```

```python
# training the word2Vec
# Train Word2Vec model
model = Word2Vec(sentences=tokenized_corpus, vector_size=100, window=5, min_count=1, workers=4)

# Save the trained model
model.save("word2vec.model")
```

```python
# test the model
# Load the model
model = Word2Vec.load("word2vec.model")

# Get vector representation of a word
vector = model.wv['machine']
print("Vector representation of 'machine':\n", vector)

# Find similar words
similar_words = model.wv.most_similar('machine', topn=3)
print("\nWords similar to 'machine':", similar_words)
```
